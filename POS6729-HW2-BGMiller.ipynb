```python
#set up while loop to go through the whole file
#initialize everything
x = 0 # this is the index of the record in data.  In python records start with 0, 1, 2, ...
current_node = data.Edge[0] # initialize current node to the first node in the file (1)
min_edges = 1 # the minimum number of edges wanted in the final node list
newdata = []  # set up an empty list
Flag = 0 #flag for if it has written to the list yet
counter = 0 # counter for each node to count the number of Edge = 1
#loop through the file
while x < len(data):
    # if node is not current node, set current node to new node
    if data.Node1[x] == current_node:   # if record x of the field Node1 = current node
        if data.Edge[x] == 1:
            counter = counter + 1
            if counter >= min_edges:
                if Flag != 1:
                    newdata.append(data.Node1[x])
                    Flag = 1
                #j = j+1
    else:
        current_node = data.Node1[x] # set the current node
        Flag = 0 #reset the flag
        counter = 0 #reset the counter
    #print(x,data.Node1[x],counter)
    x = x + 1
print('the number of nodes with >= ', min_edges, 'equals ', len(newdata),'.', 'They are ', newdata)
```
