```python
##############################################
#      
#         HW 2 = with min_edges = 1
#                        
##############################################

#set up while loop to go through the whole file
#initialize everything
x = 0 # this is the index of the record in data.  In python records start with 0, 1, 2, ...
current_node = data.Node1[0] # initialize current node to the first node in the file (1)
min_edges = 1 # the minimum number of edges wanted in the final node list
newdata = []  # set up an empty list
Flag = 0 #flag for if it has written to the list yet
counter = 0 # counter for each node to count the number of Edge = 1
# first sort file by node1, then node2 to ensure logic below works
data=data.sort_values(by=['Node1', 'Node2'])
print(data)
#loop through the file
while x < len(data):
    # if node is not current node, set current node to new node
    if data.Node1[x] == current_node:   # if record x of the field Node1 = current node
        if data.Edge[x] == 1:
            counter = counter + 1
            if counter >= min_edges:
                if Flag != 1:
                    newdata.append(data.Node1[x])
                    Flag = 1
                #j = j+1
    else:
        current_node = data.Node1[x] # set the current node
        Flag = 0 #reset the flag
        counter = 0 #reset the counter
    #print(x,data.Node1[x],counter)
    x = x + 1
print('the number of nodes (Node1 to node2) with edges >= ', min_edges, 'equals ', len(newdata),'.', 'They are ', newdata)
#next sort by node2, then node1 to uncover any hidden connections
data2=data.sort_values(by=['Node2','Node1'],ignore_index=True)
print(data2)
#rerun logic
newdata2=[]#set up second empty list
counter = 0 # counter for each node to count the number of Edge = 1
current_node = data2.Node2[0]
Flag=0
x=0
#print(current_node)
while x < len(data2):
    # if node is not current node, set current node to new node
    if data2.Node2[x] == current_node:   # if record x of the field Node1 = current node
        if data2.Edge[x] == 1:
            counter = counter + 1
            if counter >= min_edges:
                if Flag != 1:
                    newdata.append(data2.Node2[x])
                    newdata2.append(data2.Node2[x])
                    Flag = 1
    else:
        current_node = data2.Node2[x] # set the current node
        Flag = 0 #reset the flag
        counter = 0 #reset the counter
    #print(x,data2.Node2[x],counter)
    x = x + 1
print('the number of nodes (from Node2 to Node1) with edges >= ', min_edges, 'equals ', len(newdata2),'.', 'They are ', newdata2)
#Now we remove duplicates from the newdata list
mylist = list(dict.fromkeys(newdata)) 
newdata.sort()
print('sorted newdata = ',newdata)
print('')
print('','mylist= ', mylist)
print('the number of edges (in total) with edges >= ', min_edges,'equals ', len(mylist))
```
